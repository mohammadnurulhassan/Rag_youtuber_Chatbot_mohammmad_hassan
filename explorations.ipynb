## Semantic  in LanceDB

```python


def search_knowledge_base(query: str, k: int = 3):
    """
    Search LanceDB for k most relevant documents for a given query.
    """
    results_df = (
        articles.search(query)
        .limit(k)
        .to_pandas()
    )
    return results_df

search_knowledge_base("vad r dbt?", k=5)[["doc_id", "filename", "content"]]
```

### Call  RAG agent directly

```python



import asyncio
from backend.rag import rag_agent

async def ask_rag(question: str):
    result = await rag_agent.run(question)
    resp= result.output
   
    print("Question:", question)
    print("-" * 50)
    print("Answer:\n", resp.answer)
    print("\nSource filename:", resp.filename)
    print("Source path:", resp.filepath)

   
await ask_rag("what is dbt and why we use in modern data stack?") 
```
